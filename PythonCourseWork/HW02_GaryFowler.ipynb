### Plotting experimental data


Given a file, sunspots.txt, which contains the observed number of sunspots for each month since January 1749.
a)write a program that reads in the data and makes a graph of the sunspots as a function of time.
b)Modify the program to display only the first 1000 data points on the graph.
c)Modify your program further to calculate and plot the running average of the data defined:
Yk = 1 / (2r + 1) (sum(Yk+m) from m = -r to r)
where r = 5 in this case and Yk are the sunspot numbers. Have the program plot the data and the running average, again with 1000 data points.

```python
## @author: Gary Fowler

"""
@data: data drawn from sunspots.txt
@x: the first column, holds an index
@y: the second column, holds the number of sunspots
@x2, y2: sliced versions of the created arrays x and y
@r: given number for the running average formula
@m: part of the formula, also the bottom bound of the sum
@s: the running sum to calculate the average
@formula: the formula used to calculate the average

in this program we will read data from a file and manipulate it in many ways

"""
import numpy as np
import matplotlib.pyplot as plt

#part a
data = np.loadtxt("sunspots.txt",float)
fig = plt.figure()
ax = fig.add_subplot(111)

x = data[:,0]
y = data[:,1]
ax.set_xlabel('Time in Months')
ax.set_ylabel('Number of Sunspots')

plt.plot(x,y)
plt.show()

#part b
x2 = x[0:1000]
y2 = y[0:1000]
fig2 = plt.figure()
ax2 = fig2.add_subplot(111)

ax2.set_xlabel('Time in Months')
ax2.set_ylabel('Number of Sunspots')
    
plt.plot(x2,y2)
plt.show()

#part c
r = 5
m = -r

#the running sum
s = y2 + m

#since we go from -r to r, we can just double r
for k in range(1,2*r):
    s += y2 + m
    
#the average    
formula = (1 / (2*r + 1)) * s

plt.plot(x2,y2,formula,y2)
plt.show()
```

### Curve plotting


Although the plot function is designed primarily for plotting standard xy graphs, it can be adapted for other kinds of plotting.
a)Make a plot of the so-called deltoid curve, which is defined by the equations:
x = 2cos(theta) + cos(2theta), y = 2sin(theta) - sin(2theta)
where 0 <= theta <= 2pi. Take a set values of theta between zero and 2pi, calculate x and y, then plot y as a function of x.
b)Taking this approach a step farther, make a polar plot r = f(theta) by converting r and theta to Cartesian coordinates using the eguations:
x = rcos(theta), y = rsin(theta).
Use this method to plot the Galilean spiral r = theta^2 for 0 <= theta <= 10pi.
c)Using this same method make a polar plot of Fey's function:
r = e^cos(theta) - 2cos(4theta) + sin^5(theta/12) in the range 0 <= theta <= 24pi.

```python
## @author: Gary Fowler

"""
@steps: variable to easily change number of steps within the program
@theta: holds theta value, in this instance 0 to 2pi
@xpoints: the formula for deltoid curve x values
@ypoints: the formula for deltoid curve y values
@r: used for polar coordinates, a given value
@m: part of the formula, also the bottom bound of the sum
@s: the running sum to calculate the average
@formula: the formula used to calculate the average
@xpolar, ypolar: the polar coordinates conversion formulas

in this program we will plot the deltoid curve, Galilean spiral, and Fey's function

"""
import numpy as np
import matplotlib.pyplot as plt

steps = 1000
theta = np.linspace(0,2*np.pi,steps)

#part a
xpoints = 2*np.cos(theta) + np.cos(2*theta)
ypoints = 2*np.sin(theta) - np.sin(2*theta)

plt.plot(xpoints,ypoints)
plt.show()

#part b Galilean spiral
theta = np.linspace(0,10*np.pi,steps)
r = theta**2

xpolar = r*np.cos(theta)
ypolar = r*np.sin(theta)

plt.plot(xpolar,ypolar)
plt.show()

#part c Fey's function
theta = np.linspace(0,24*np.pi,steps)
r = np.exp(np.cos(theta)) - 2*np.cos(4*theta) + np.sin(theta / 12)**5

xpolar = r*np.cos(theta)
ypolar = r*np.sin(theta)

plt.plot(xpolar,ypolar)
plt.show()
```
